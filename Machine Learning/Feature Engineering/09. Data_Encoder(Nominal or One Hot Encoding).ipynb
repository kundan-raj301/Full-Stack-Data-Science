```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
# Create an o=instance of One Hot Encoder
encoder = OneHotEncoder()
```

```python
encoder.fit_transform(df[['day']]).toarray() #alphabetically sort
```

```python
#  fit the encoder to the dataframe and transform thr categorical variable
encoded = encoder.fit_transform(df[['day']])
```

```python
import pandas as pd
encoded_df = pd.DataFrame(encoded.toarray(),columns = encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
pd.concat([df,encoded_df],axis=1)
```
