# Q1. What is an API? Give an example, where an API is used in real life.

Using a set of definitions and protocols, APIs are techniques that let two software components communicate with one another. 
For Example, the software system of the weather bureau contains daily weather information. Our phone's weather app "talks" to this system via APIs to provide us with daily weather updates.

# Q2. Give advantages and disadvantages of using API.
Advantages of APIs:-

1.Efficiency: Compared to human-produced outputs in an organisation, API produces results that are more effective, dependable, and timely.
2.Flexible service delivery: API offers quick and flexible service delivery based on the needs of developers.

3.Automation: API produces better and more automated solutions since it uses robotic computers rather than humans.
4.New functionality: As a result of leveraging APIs, developers have discovered new resources and capabilities for API trading.

Disadvantages of APIs :-
1.Cost: Developing and implementing API is costly at times and requires high maintenance and support from developers.
2.Security issues: Using API adds another layer of surface which is then prone to attacks, and hence the security risk problem is common in APIs.

SOAP - Architecture:-



1.Because it is built on XML, SOAP is a lightweight data transfer system.
2.SOAP was created to be platform and OS neutral.

3.It is based on HTTP, which is present in the majority of computers.
4.It is recommended by the W3 consortium, which functions as the Web's governing body.
5.Web services and application programming interfaces (APIs) are the principal uses of SOAP.
